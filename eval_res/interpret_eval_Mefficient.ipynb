```python
import json
import pandas as pd
```

```python
with open("validation_metrics_mefficient_no_freeze.json", "r") as f:
    mefficient_data = json.load(f)

with open("validation_metrics.json", "r") as f:
    baseline_data = json.load(f)
    baseline_data = baseline_data["per_class_average_precision"]

    



```

```python
label_map = pd.read_excel("../imat_data/relabel-split.xlsx", engine='openpyxl')
```

```python
label_map.dtypes
```

```python
grouped_mefficient = label_map.groupby("taskName")

for category, group in grouped_mefficient:
    for key, value in mefficient_data.items():
        if key.startswith(category) and "per" in key:
            data_curr = value
    print(f"Category: {category}")
    for index, row in group.iterrows():
        
        print(f"Label: {row['labelName']}, Mefficient: {data_curr[row['group_label']]}, Baseline: {baseline_data[row['labelId_new']]}")
```
